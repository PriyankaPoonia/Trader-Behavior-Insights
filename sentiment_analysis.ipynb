```python
import pandas as pd

# Load data
sentiment = pd.read_csv("fear_greed_index.csv")
trades = pd.read_csv("historical_data.csv")

# Preprocess
sentiment.columns = sentiment.columns.str.strip()
trades.columns = trades.columns.str.strip()
trades['date'] = pd.to_datetime(trades['Timestamp IST'], format="%d-%m-%Y %H:%M").dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

# Merge & classify sentiment
merged = trades.merge(sentiment[['date', 'classification']], on='date', how='left')
merged['classification'] = merged['classification'].fillna('')
merged['sentiment'] = merged['classification'].apply(lambda x: 'Greed' if 'Greed' in x else 'Fear')

# Group and analyze
result = merged.groupby(['sentiment'])['Closed PnL'].agg(['count', 'sum', 'mean']).reset_index()
print(result)
```
